```python
import os
import getpass
```

## Pegando API do HugginFace

```python
os.environ["HUGGINGFACEHUB_API_TOKEN"] = getpass.getpass()
```

## Fazendo as importações das bibliotecas

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
```

### Criação do template
O template é o prompt onde é descrito o que o bot tem que fazer.

```python
template = """Você é o atendente virtual da empresa "Gesso Galo", especialista em serviços de gesso e drywall.
Escreva uma mensagem profissional, educada e comercial respondendo ao cliente.

O cliente solicitou orçamento para:
Serviço: {servico}
Metragem: {metragem} metros quadrados

Forneça uma resposta cordial, agradecendo o contato e informando que um especialista entrará em contato em breve para alinhar os detalhes técnicos.
Responda diretamente em português de forma clara.
Para ficar mais profissional adicione o nome do especialista no final da mensagem (Pedro) e o número para contato (123456789).
"""
```

### Criação da LLM
Estamos usando o modelo Qwen2.5, que tem um desempenho bom (por ser um modelo open source)
* temperature: Temperatura para 0.3, para manter a imaginação equilibrada.
* max_new_tokens: Limites de tokens para evitar textos grandes e uso excessívo de tokens.
* task: Informando que queremos o modelo de chat.

```python
llm_endpoint = HuggingFaceEndpoint(
    repo_id="Qwen/Qwen2.5-7B-Instruct",
    task="conversational",
    max_new_tokens=512,
    temperature=0.3,
)
```

### Montando o sistema para o Langchain
Como o endpoint não fica estruturado para o modelo que o Langchain pede (SystemMessage, HumanMessage, AIMessage), o ChatHugginFace faz esse trabalho de "tradução".

```python
chat_model = ChatHuggingFace(llm=llm_endpoint)
```

### Injeção de valores
O ChatPromptTemplate faz uma injeção de valores no comandos dentro de chaves, entendendo como um input.

```python
prompt = ChatPromptTemplate.from_template(template)
```

### Limpeza de saída
O StrOutputParser faz a limpeza da saída, entregando um texto limpo, ao invés de um texto com códigos.

```python
output_parser = StrOutputParser()
```

## Langchain
Aqui é a fórmula principal do Langchain, onde ele vai fazer o encadeamento do código, informando o passo a passo.

```python
gesso_galo_chain = prompt | chat_model | output_parser
```

```python
dados_cliente = {
    "servico": "Sanca Aberta com Iluminação LED",
    "metragem": "35"
}
```

## Resposta da LLM
O invoke faz a chamada do modelo e passa como parametro o input do usário.

```python
resposta_final = gesso_galo_chain.invoke(dados_cliente)
print(resposta_final)
```
